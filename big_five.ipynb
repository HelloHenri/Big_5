```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

#importando mais bibliotecas
```

```python
kmeans = KMeans()
visualizador = KElbowVisualizer(kmeans, k=(2,10))

#k é o número de grupos que o kmeans faça agrupamento de dados
```

```python
amostra_dados = dados.sample(n=9000)

#filtrando uma amostra da base da dados para 9000 para diminuir a demora para processar
```

```python
visualizador.fit(amostra_dados)
visualizador.poof()

#fit para as variações do metodo k anteriormente atribuido, e poof mostra graficamente.
```

# ***Com a execução do código acima, podemos notar que o número de agrupamentos que "k" é mais compatível é 5, ou seja, 5 agrupamentos é o melhor número para essa base de dados***

```python
kmeans = KMeans(n_clusters=5)
k_fit = kmeans.fit(dados)

#especificamos o n° de grupos = 5 para n_clusters, e amarzenamos os dados de kmeans.fit para a variavek k_fit
```

```python
predicao = k_fit.labels_
dados['Agrupamentos'] = predicao

#variavel predicao para armazenar k_fit.labels e passar os valores de predicao para um novo atributo "Agrupamentos"
```

```python
# Função para categorizar os traços
def categoriza(x):
    if x >= 4:
        return 'Alta'
    elif x >= 3:
        return 'Media'
    else:
        return 'Baixa'

# Criar categorias usando as colunas das médias que você já calculou
dados['categoria_ext'] = dados['extroversao'].apply(categoriza)
dados['categoria_neu'] = dados['neuroticismo'].apply(categoriza)
dados['categoria_agr'] = dados['amabilidade'].apply(categoriza)
dados['categoria_csn'] = dados['conscienciosidade'].apply(categoriza)
dados['categoria_opn'] = dados['abertura'].apply(categoriza)

# Importar bibliotecas para plotar os gráficos
import seaborn as sns
import matplotlib.pyplot as plt

# Dicionário para facilitar os plots
traços = {
    'Extroversão': 'categoria_ext',
    'Neuroticismo': 'categoria_neu',
    'Amabilidade': 'categoria_agr',
    'Conscienciosidade': 'categoria_csn',
    'Abertura': 'categoria_opn'
}

# Criar os gráficos de distribuição das categorias
plt.figure(figsize=(15,10))
for i, (nome, coluna) in enumerate(traços.items(), 1):
    plt.subplot(2, 3, i)
    sns.countplot(x=coluna, data=dados, order=['Baixa', 'Media', 'Alta'])
    plt.title(f'Distribuição da {nome}')
plt.tight_layout()
plt.show()
```

Verificamos quantas pessoas nas seguintes personalidades separando por categorias: altas (>=4), medias (>=3) e baixas (<3)
